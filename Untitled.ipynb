```python
path = "Et dukkehjem - tekst.md"

with open(path,'r',encoding = 'utf-8') as f:
    text = f.read()

personer = ["nora","torvald","helmer","fru linde","krogstad"]

print("done")
```

```python
while(True):
    place = text.find('|')
    
    text = text[:place] + '\n\n' + text[place + 1:]
    
    if(place == -1):
        break


        
#print(text)
    
    
```

```python
import re

#text = "nora: en person. torvald: en anden person"

pos = {}

for person in personer:
    pos[person] = [m.start() for m in re.finditer(person + ':', text)]
    
print(pos.keys())
```

```python
extra_chars = 0

insert_string = "\n\n---------------\n\n"

for person in personer:
    for line_start in pos[person]:
        text = text[:line_start] + insert_string + text[line_start:]
        extra_chars += len(insert_string)

print(text)
```
